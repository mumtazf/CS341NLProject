```python
import pandas as pd
import numpy as np
```

```python
laptops = pd.read_csv("data\laptops.csv")
print(laptops.shape)
```

```python
## Does the dataset have any NAs?
na_values = laptops.isna().sum()
na_values
```

```python
print(laptops['price (usd)'].describe())
```

```python
print(laptops['ram_memory'].describe())
```

```python
laptop_groups = laptops.groupby('Model')
laptop_counts = laptop_groups.size()
print(laptop_counts)
ram_memory = laptop_counts.index.tolist()

print(ram_memory)
```

```python
def parseBrandName(str):
    result = str.split("(")
    return result[0]

laptops['Model'] = laptops['Model'].apply(lambda x : parseBrandName(x))
```

```python
laptop_groups = laptops.groupby('brand')
laptop_counts = laptop_groups.size()
print(laptop_counts)
brand_names = laptop_counts.index.tolist()

print(brand_names)
```

```python
def func_price(price):
    if price < 700:
        return 'low'
    elif price >= 700 and price <= 1000:
        return 'medium'
    else:
        return 'high'

laptops['price_range'] = laptops['price (usd)'].apply(lambda x : func_price(x))
    
```

```python
laptops.to_csv("laptops.csv")
```
